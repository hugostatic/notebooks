```
nbGrenouilles = int(input())
listeGrenouilles = [0 for x in range(nbGrenouilles)]
victoires = [0 for x in range(nbGrenouilles)]

# print(listeGrenouilles)

nbTours = int(input())
for i in range(nbTours):
    saut = input().split(" ")
    grenouille = int(saut[0])
    distanceSaut = int(saut[1])
    listeGrenouilles[grenouille-1] += distanceSaut 
    victoires[listeGrenouilles.index(max(listeGrenouilles))] += 1
print(victoires.index(max(victoires))+1)
```

#Carré magique
http://www.france-ioi.org/algo/task.php?idChapter=564&idTask=469
Un carré magique est une grille carrée dans laquelle des nombres sont placés de telle sorte que la somme des nombres de chaque colonne, chaque ligne et de chacune des deux diagonales soit la même. De plus, Le carré doit contenir une fois chaque nombre, de 1 au nombre de cases de la grille.
Ecrivez un programme qui vérifie si une grille de nombres est un carré magique.
1 <= N <= 20, où N est le nombre de lignes et de colonnes de la grille.

__Entrée__ : La première ligne de l'entrée contient un entier N : le nombre de cases du côté de la grille de nombres.
Chacune des N lignes suivantes contient N entiers séparés par des espaces : les nombres d'une ligne de la grille.

__Sortie__ : Vous devez afficher une ligne sur la sortie, contenant le mot "yes" si le carré fourni est un carré magique, et "no" sinon.

```
dim = int(input())
if dim>20:
   dim = int(input())
carre = []
# on remplit le carré
for i in range(dim):
  ligne = input().split()
  carre.append(ligne)

def somme(liste):
  s = 0
  for i in range(len(liste)):
    s += int(liste[i])
  return s

def transposee(matrice):
  transposee = [["" for j in range(len(matrice))] for i in range(len(matrice))]
  for i in range(len(matrice)):
    for j in range(len(matrice)):
      transposee[i][j] = matrice[j][i]
  return transposee

def sommeCol(matrice):
  s = 0
  for i in range(len(matrice)):
    s += int(matrice[i][i])
  return s

def sommeCol2(matrice):
  s = 0
  for i in range(len(matrice)):
    s += int(matrice[len(matrice)-1-i][i])
  return s

def testCarre(tableau):
  entiers = [i for i in range(1,len(tableau)**2+1)]
  for i in range(len(tableau)):
    for j in range(len(tableau)):
      if tableau[i].count(tableau[i][j]) != 1 or int(tableau[i][j]) not in entiers:
        return "no"
  sommeLigne = somme(tableau[0])
  for i in range(1,len(tableau)):
    #somme des termes d'une ligne ou d'une colonne 
    if somme(tableau[i]) != sommeLigne or somme(transposee(tableau)[i]) != sommeLigne or sommeCol(tableau) != sommeLigne or sommeCol2(tableau) != sommeLigne:
      return "no"
  return "yes"

print(testCarre(carre))
```
