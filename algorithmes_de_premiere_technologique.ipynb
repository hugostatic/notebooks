\- utiliser le principe d’accumulateur pour calculer une somme, un produit

```python
def somme(début,fin): # Calcule la somme des nombres de début à fin
    s = 0             # s est l'accumulateur de la somme calculée
    for n in range(début,fin+1):
        s = s + n
    return s

somme(4,12)
```

```python
def produit(début,fin): # Calcule la somme des nombres de début à fin
    p = 1               # p est l'accumulateur de la somme calculée
    for n in range(début,fin+1):
        p = p * n
    return p

produit(4,12)
```

## Calculer un terme de rang donné d’une suite, une somme finie de termes.

```python
# Par la définition récurrente

def terme_suivant(u):
    return 0.9*u+10

u_0 = 2019  # premier terme

def calculer_terme(rang):
    u = u_0
    for i in range(rang):
        u = terme_suivant(u)
    return u

print("Le terme de rang 10 est :",calculer_terme(10))
```

```python
def somme_jusqu_au_rang(rang):
    u = u_0
    somme = u_0
    for i in range(rang):
        u = terme_suivant(u)
        somme = somme + u
    return somme

print ("La somme des dix premiers termes est :",somme_jusqu_au_rang(9))
```

```python
def somme_partielle(rang_initial,rang_final):
    u_initial = calculer_terme(rang_initial)
    u = u_initial
    somme = u_initial
    for i in range(rang_initial+1,rang_final+1): # du terme suivant le terme initial jusqu'au terme final inclus
        u = terme_suivant(u)
        somme = somme + u
    return somme

print ("La somme de u_6 à u_10 est :",somme_partielle(6,10))
```

## Déterminer une liste de termes d’une suite et les représenter.

```python
# Par la définition récurrente

def terme_suivant(u):
    return 0.9*u+10

u_0 = 2019  # premier terme

def liste_termes(rang) : # rang désigne le rang du dernier terme
    u = u_0
    liste = [u]
    for i in range(rang): 
        u = terme_suivant(u)
        liste = liste + [u]
    return liste

print("Les dix premiers termes de la liste sont :",liste_termes(9))
```

```python
import matplotlib.pyplot as plt

def representation_graphique(suite):
    n = len(suite)                     # détermine le nombre de termes dans la suite
    abscisses = list(range(n))         # crée la liste de 0 à n-1
    ordonnées = suite
    plt.plot(abscisses,ordonnées,"b.")
    plt.show()
    return

suite = liste_termes(50)
representation_graphique(suite)
```

## Déterminer le rang à partir duquel les termes d'une suite sont supérieurs ou inférieurs à un seuil donné, ou aux termes de même rang d'une autre suite.

**Déterminer le rang à partir duquel les termes d'une suite sont supérieurs ou inférieurs à un seuil donné**

```python
# Par la définition récurrente

def terme_suivant(u):
    return 0.9*u+10

u_0 = 2019  # premier terme

def rang(seuil):
    u = u_0
    n = 0
    while u >= seuil:    # Attendu que la suite considérée est décroissante
        u = terme_suivant(u)
        n = n + 1
    return n

print("Le premier terme inférieur à 500 est de rang :",rang(500))
```

## Calculer une valeur approchée d’une solution d’une équation par balayage.

```python
def f(x):
    return 3*x**2-2*x-1

def resolution(f,y,a,b,h): # Résolution par balayage de l'équation : f(x) = 0 sur l'intervalle [a,b] avec un pas h
    x = a
    
    while (f(x)-y)*(f(x+h)-y) > 0 and x < b : # Balayage
        x = x + h
    
    if x > b : # Cas où aucune solution n'est trouvée
        x = "Pas de solution"
    
    return x

resolution(f,4,0,10,0.1)
```

**Remarque:** Il peut être intéressant de proposer une illustration graphique de la situation.

## Simuler des échantillons de taille <i>n</i> d’une loi de Bernoulli à partir d’un générateur de nombres aléatoires entre 0 et 1.

```python
from random import random

def loi_Bernouilli(p):
    if random() <= p :
        issue = "Réussite"
    else :
        issue = "Echec"
    return issue

def echantillonner(n,p): # Création d'un échantillon de taille n d'une loi de Bernouilli de paramètre p
    echantillon = []
    for i in range(n):
        if loi_Bernouilli(p) == "Réussite" :
            resultat = 1
        else:
            resultat = 0
        echantillon.append(resultat)
    return echantillon

echantillon = echantillonner(500,0.35)
print("Echantillon de taille 2019 avec p = 0,35 :")
print(echantillon)
```

## Représenter par un histogramme ou par un nuage de points les fréquences observées des 1 dans N échantillons de taille n d’une loi de Bernoulli.

```python
def liste_fréquences(N,n,p): # Création de la liste des fréquences observées des 1 
    L=[]                     # dans N échantillons de taille n d'une loi de Bernouilli de paramètre p
    for i in range(N):
        echantillon = echantillonner(n,p)
        frequence = sum(echantillon)/n
        L.append(frequence)
    return L

frequences = liste_fréquences(100,500,0.35) # 100 échantillons de taille 500 d'une loi de Bernouilli de paramètre p = 0.35
print("Liste des fréquences dans 100 échantillons de taille 500 avec p = 0.35 :")
print(frequences)
```

```python
import matplotlib.pyplot as plt

def histogramme(liste):
    plt.title("Histogramme représentant les fréquences")
    plt.xlabel("Valeurs")
    plt.ylabel("Fréquences")
    plt.hist(liste)
    return

histogramme(frequences)
```

```python
def nuage_de_points(liste):
    abscisses = list(range(len(liste)))
    plt.title("Nuage de points représentant les fréquences")
    plt.xlabel('Echantillon')
    plt.ylabel('Fréquence')
    plt.scatter(abscisses,liste)
    plt.show()
    return

nuage_de_points(frequences)
```

## Compter le nombre de valeurs situées dans un intervalle de la forme [p - ks ; p + ks] pour k ∈ {1;2;3}.

```python
p = 0.35
n = 500

def moyenne(liste):
    return sum(liste) / len(liste)

def variance(liste):
    m = moyenne(liste)
    return moyenne([(x-m)**2 for x in liste])

def ecartype(liste):
    return variance(liste)**0.5

s = ecartype(frequences)

for k in [1,2,3]:
    nombre_valeurs = 0
    for f in frequences :
        if p-k*s <= f and f <= p+k*s :
            nombre_valeurs += 1
    print("Le nombre de fréquences comprises entre p -",k,"* s =",p-k*s,"et p +",k,"* s =",p+k*s,"est",nombre_valeurs)
    
```

## Construire différents points d’une approximation de courbe intégrale par la méthode d’Euler.

```python
import matplotlib.pyplot as plt

def approximation_euler(f,a,b,y_0,n): # Approximation de la courbe intégrale de f sur [a,b] avec F(a)=y_0 en n étapes
    x,y = a,y_0
    abscisses,ordonnees = [x],[y]
    h = (b-a)/n
    for i in range(n):
        y = y + h * f(x)
        x = x + h
        abscisses.append(x)
        ordonnees.append(y)
    plt.plot(abscisses,ordonnees,'rx')
    plt.show()
    return 
```

```python
def f(x):
    return x

approximation_euler(f,0,10,1,10)
```

```python
def f(x):
    return 1/x

approximation_euler(f,1,10,0,10)
```
