Trouver le plus petit nombre premier supérieur à 100 qui s'écrit uniquement avec des 1 et/ou des 7 : 

```
def is_prime(x):
  if x<2:
    return False
  for i in range(2,x):
    if x%i == 0:
      return False
  return True  
```

```
is_prime(21)
```

```
is_prime(13)
```

```
def test17(n):
  chaine=str(n)
  for i in range(len(chaine)):
    if chaine[i] != "1" and chaine[i] != "7":
      return False
  return True
```
