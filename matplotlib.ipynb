```
import numpy as np
import matplotlib.pyplot as plt

#————————————————plotting area
x1=-10
x2=140  
y1=-10
y2=90 
plt.axis([x1,x2,y1,y2])
plt.axis('on')                      #cette instruction n'est pas indispensable mais elle permet d'effacer les axes facilement

#——————————————————grid
plt.grid(True,color='#11B5E4')
plt.title('Tick Mark Sample')       #Titre

#————————————————tick marks         
xmin=x1
xmax=x2
dx=10 
ymin=y1 
ymax=y2  
dy=5
plt.xticks(np.arange(xmin, xmax, dx)) 
plt.yticks(np.arange(ymin, ymax, dy))

plt.show()                          #cette instruction n'est pas indispensable
```

```
import numpy as np
import matplotlib.pyplot as plt

#————————————————plotting area
x1=-5
x2=15  
y1=-15
y2=5
plt.axis([x1,x2,y1,y2])
plt.axis('on')                      #cette instruction n'est pas indispensable mais elle permet d'effacer les axes facilement

#——————————————————grid
plt.grid(True,color='#11B5E4')
plt.title('CUSTOM GRID')       #Titre
plt.xlabel('this is the x axis') 
plt.ylabel('this is the y axis')
#————————————————tick marks         
xmin=x1
xmax=x2
dx=.5 
ymin=y1 
ymax=y2  
dy=.5
for x in np.arange(x1,x2,dx):
  for y in np.arange(y1,y2,dy):
    plt.scatter(x,y,s=1,color='#5C0029')    # s=1 correspond à la taille du point

plt.show()                          #cette instruction n'est pas indispensable
```

```
from pylab import *

def f(t):
    s1 = cos(2*pi*t)
    e1 = exp(-t)
    return multiply(s1, e1)

t1 = arange(0.0, 5.0, 0.1)
t2 = arange(0.0, 5.0, 0.02)
t3 = arange(0.0, 2.0, 0.01)

subplot(211)
l = plot(t1, f(t1), 'bo', t2, f(t2), 'k--', markerfacecolor='green')
grid(True)
title('2 graphiques ensemble')
ylabel('des oscillations')

subplot(212)
plot(t3, cos(2*pi*t3), 'r.')
grid(True)
xlabel('temps (s)')
ylabel('encore des oscillations')
show()
```

```
from pylab import *

couleurs = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

x = [-1, 0, 1, 2]
y = [3, 2, 4, 1]

for c in couleurs:
    i = couleurs.index(c) + 1
    subplot(2, 4, i, facecolor='#FCF5C7')  # ajout de couleur de fond
    xticks([])                    # aucune etiquette sur l'axe des x
    yticks([])                    # aucune etiquette sur l'axe des y
    plot(x, y, c+'-')

show()
```

```
from pylab import *

style = ['-', '--', '-.', ':']

x = [-1, 0, 1, 2]
y = [3, 2, 4, 1]

for s in style:
    i = style.index(s) + 1
    subplot(1, 4, i, facecolor='#EEB8FE')     # code couleur html
    xticks([])
    yticks([])
    plot(x, y, 'b'+s)
    axis([-2,3,0,5])

show()
```

```
from pylab import *

marqueurs = ['o', '+', '.', 'x', '*', '^']

x = [-1, 0, 1, 2]
y = [3, 2, 4, 1]

for m in marqueurs:
    i = marqueurs.index(m) + 1
    subplot(2, 3, i, facecolor='#FCF5C7')     
    xticks([])
    yticks([])
    plot(x, y, 'g'+m)
    axis([-2,3,0,5])

show()
```

```
from pylab import *

x=arange(-6.28,6.28+0.05,0.05)

plot(x, exp(-x**2)+.4, '#D36582', label='sinus')
plot(x, cos(x), 'r--', label='cosinus')

axis([-7,7,-1.5,1.5])
xticks(arange(-6,6+0.1,0.2), '')
axhline(color='k')  # axe des x
axvline(color='k')  # axe des y
legend()            # reprend les labels des plots

show()
```

```
from pylab import *

niveaux = ['Seconde', 'Premiere', 'Terminale']
couleurs = ['#A03E99', '#F1BF98', '#F6828C']
repartition = [125, 210, 170]

pie(repartition, labels=niveaux, colors=couleurs, autopct='%1.1f%%')
axis('equal')

show()
```
