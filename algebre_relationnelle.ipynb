# Concept de relation
## L'objet relation :

**relation** = tableau où l'on range des objets de même nature

**objet** = ligne du tableau = tuple constitué d'un **identifiant** unique et de valeurs.

**attribut** = colonne du tableau. L'ensemble des attributs est le **schéma** de la relation. L'ensemble des valeurs que peuvent prendre les attributs est le **domaine**  Parfois pour un tuple la valeur d'un attribut n'existe pas (**None** en Python).

**La clé**= groupe d'attributs minimum déterminant un tuple unique. ici clé = ID. 

Une clé ne doit pas contenir de doublon... En Python, pour détecter les doublons il faut employer la méthode drop_duplicates() de la librairie Pandas.

Il y a souvent plusieurs clés possibles (candidates) et on en choisit une  (En connaître au moins une est très important : elles déterminent la nature des objets représentés.)

clé = id : l'objet est une pomme 
clé = id - propriétaire : l'objet est une association...

Pourquoi ? Parce qu’elles déterminent la nature des objets que vous représentez dans une table.):

la clé **primaire** : PK. Si aucune ne convient on crée une clé **artificielle**. Une clé primaire ne doit pas contenir de valeur NULL

## Liens entre relations :

**clés étrangères** FK

relation Pomme - relation Variété
Dans la schéma de la relation Pomme il y a un attribut qui correspond à la clé primaire de la Relation variété.

La clé étrangère nom_variété **référence** la clé primaire libellé de la relation Variété.

## Éviter la redondance 
dans le stockage des données. (par forcement dans leur questionnement)

Dans une relation, si un attribut A (prix au kilo) dépend uniquement d'un groupe d'attributs G (nom_variété) (et que ce groupe d'attributs n'est pas une clé candidate), alors il est possible de créer une nouvelle relation qui contiendra les attributs A et G (clé candidate pour la nouvelle relation).

Pour savoir si A dépend uniquement de G  : "En ne connaissant que G, et en ayant à disposition la relation (Pomme), puis-je trouver de manière certaine A ?"

Il faut s'assurer que G soit minimal (c'est-à-dire que l'on ne puisse pas enlever d'attribut au groupe G sans casser la dépendance entre A et G).

## Opérateurs : projection et restriction
**projection** = suppression d'attributs donc de colonnes 

pomme_proj = Projection (pomme, masse, couleur ) 

**restriction** = suppression d'objets donc de lignes  en posant une condition sur l'attribut (diamètre_pomme > 80mm)

pomme_restr = Restriction (pomme, masse > 141 g)

N.B. restriction avant projection ;)

## La jointure
### jointure interne
Coller par exemple la relation pomme et la relation variété... condition de la jointure : pomme.nom_variete = variete.libelle

Jointure (pomme, variété, pomme.nom_variete = variete.libelle )

**Une jointure est un produit cartésien suivi d'une restriction selon la condition de la jointure...**

Possibilité de faire la jointure sur plusieurs colonnes...

### jointure externe

## L'agrégation
On utilise l'agrégation lorsque l'on veut calculer un résultat qui porte sur plusieurs lignes d'une relation (ex. masse moyenne des pommes d'une même couleur).

Pour cela on a besoin de 
* un **groupe d'attributs de partitionnement** : on va par exemple créer des **agrégats** de pommes ayant la même couleur... Partionnement sur plusieurs attributs :
![Texte alternatif…](https://user.oc-static.com/upload/2017/06/26/14984793735677_PARTITIONNEMENT.png)
* une (ou des) **fonction(s) d'agrégation** 

Généralement elle recoit une liste et renvoie une valeur unique...
